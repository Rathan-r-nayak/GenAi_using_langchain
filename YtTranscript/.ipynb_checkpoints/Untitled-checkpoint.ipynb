## Document Loading

```python
def combine_texts(transcript):
    li = []
    for snip in transcript.snippets:
        li.append(snip.text)
    return " ".join(li)
```

```python
video_id = "K6KkGASvCnY"
languages = []
text = ""
try:
    api = YouTubeTranscriptApi()
    # Step 1: list all transcripts
    transcript_list = api.list(video_id)

    for t in transcript_list:
        languages.append(t.language_code)
    # print(languages)
         
    if "en" in languages:
        print("English language")
        transcript = transcript_list.find_transcript(["en"]).fetch()
        text = combine_texts(transcript)
    else:
        non_eng_transcript = transcript_list.find_transcript(languages).fetch()
        text = combine_texts(non_eng_transcript)
        # translator = Translator()|
        print(text)
        # translator.translate(text, src=non_eng_transcript.language, dest="en")
        # print(non_eng_transcript.language)
        
        # for snip in non_eng_transcript.snippets:
        #     text.append(translator.translate(snip.text, src=non_eng_transcript.language, dest="en"))
        

        
except TranscriptsDisabled:
    print("Transcripts are disabled for this video.")
except NoTranscriptFound:
    print("No transcript found for this video.")
except Exception as e:
    print("Error:", e)


# text
```
