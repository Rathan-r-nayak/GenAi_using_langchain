```python
from langchain_core.tools import tool
```

```python
@tool
def multiply(a:int,b:int)->int:
    """multiplication of 2 numbers"""
    return a*b
```

```python
multiply.invoke({'a':5,'b':4})
```

```python
print(multiply.name)
print(multiply.args)
print(multiply.description)
```

```python
multiply.args_schema.model_json_schema()
```

```python
multiply.name = "add"
```
