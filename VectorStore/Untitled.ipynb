```python
from langchain_community.vectorstores import Chroma
from langchain_community.embeddings import HuggingFaceEmbeddings
```

```python
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

vector_db = Chroma(
    persist_directory= "/home/rathan-r-nayak/Rathan/Machine Learning/machine-leaning/langchain_models/VectorStore/chroma_store",
    embedding_function=embeddings
)

print("Vector DB loaded!")
```

```python
vector_db
```

```python
query = "which programming language is used in artificial intelligence"

results = vector_db.similarity_search(query=query, k=3)
print(results)

for i, doc in enumerate(results):
    print(f"--- Result {i+1} ---")
    print(doc.page_content)
```
